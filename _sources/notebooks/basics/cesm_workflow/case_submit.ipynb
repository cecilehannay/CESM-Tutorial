# Case: case.submit

Having run create_newcase in the CIME Scripts Directory, and then the case.setup and case.build scripts in the CASE Directory, CESM is now ready to run in the Build/Run Directory. To run CESM the case.submit script must be run from the CASE Directory. The case.submit script performs many tasks that combine to perform all the steps required to run CESM. These tasks include:

- Step 1. Check Archive and Run options
- Step 2. Check if Namelists need to be rebuilt
- Step 3. Check Input Data
- Step 4. Submits the case.run script to the batch job scheduler
- Step 5. Submits the case.st_archive script dependent on the successful completion of case.run

The figure below shows the flow of operations and data from the CASE Directory to the Build/Run Directory and then to the Archive Directory.


# Take-away points

- case.submit coordinates all the tasks of running CESM.
- case.run is submitted to the batch job scheduler.
- CESM runs in the Build/Run Directory.
- case.st_archive runs after case.run completes successfully.
- The history files are move to the Archive Directory after the model runs successfully.
- If required CESM will resubmit itself to continue running after archiving
